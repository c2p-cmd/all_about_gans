```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fish = np.load("quick_draw_data/fish.npy")
print(fish.shape)
plt.figure(figsize=(10, 2))
for i in range(5):
    plt.subplot(1, 5, i + 1)
    plt.imshow((fish[i] / 255.0).reshape(28, 28), cmap='gray')
    plt.axis('off')
plt.show()
```
